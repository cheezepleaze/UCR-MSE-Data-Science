### Question 1: Conceptual (Machine Learning)

#### Explain in your own words - in about 5-8 sentences.

*how a decision tree makes predictions? why decision trees are prone to overfitting? what techniques (pruning, depth limitation, minimum sampling per leaf) can help prevent overfitting and improve generalization?*


A decision tree splits the data subsets into decision nodes and leaf nodes based on features and thresholds. Each decision node is a decision/rule, leading to a predicted value. The decision/outcome is repeated at each node until it reaches a final prediction at a leaf node.

Decision trees are prone to overfitting because they can grow large/deep, splitting data until all training data is accounted for including noise, outliers, and dataset-specific patterns. This overfits the data by focusing too specifically on the dataset instead of being able to generalize on non-training data. Techniques like pruning, depth limitation, and setting minimum samples per leaf/split improves generalization by setting contraints that remove unnecesary branches that don't contribute performance, decrease variance, and prevent the model from including noise.

### Question 2 - Programming (Statistics)

```python
# 1) sampling simulation

# params
samples = 10000
n = 30
u = 50
sd = 10

# sampling
data = np.random.normal(u, sd, (samples, n))

print(f"1) {data}")
```

```python
# 2) mean calculation

mean_set = data.mean(axis = 1)
mean_list = mean_set.tolist()

# printed only up to 10 to avoid having to see all 10k on github preview
print(f"2) {mean_list[:10]} ...")
```

```python
# 3) plotting

plt.figure(figsize = (5, 5))
plt.style.use('ggplot')
plt.xlim(39, 61)
plt.xticks(np.arange(40, 61, 2))
plt.grid(True, which = "both", linestyle = "--", linewidth = 0.5, alpha = 0.5)

# bin size set with sturges' rule - results to 15
binsize = int(np.ceil(np.log2(samples) + 1))
plt.hist(mean_list, bins =  binsize, histtype = "stepfilled", alpha = 0.75, label = "Sample Means")
plt.axvline(u, linestyle = "dashed", linewidth = 2, color = "#363537", label = "True Population Mean")

plt.xlabel("Mean")
plt.ylabel("# of Sample Sets")
plt.title("Sampling and Distribution of the Mean")
plt.legend(fontsize = 8);
```

```python
# 4) mean and sd of the sampling distribution

u_dist = mean_set.mean()
sd_dist = mean_set.std(ddof = 1)

print(f"4)\nSampling Distribution Mean: {u_dist}\nSampling Distribution Standard Deviation: {sd_dist}")
```
