### Question 2 - Programming (Statistics)

```python
# 1) sampling simulation

# params
samples = 10000
n = 30
u = 50
sd = 10

# sampling
data = np.random.normal(u, sd, (samples, n))

print(f"1) {data}")
```

```python
# 2) mean calculation

mean_set = data.mean(axis = 1)
mean_list = mean_set.tolist()

print(f"2) {mean_list}")
```

```python
# 3) plotting

plt.figure(figsize = (5, 5))
plt.style.use('ggplot')
plt.xlim(39, 61)
plt.xticks(np.arange(40, 61, 2))
plt.grid(True, which = "both", linestyle = "--", linewidth = 0.5, alpha = 0.5)

# bin size set with sturges' rule - results to 15
binsize = int(np.ceil(np.log2(samples) + 1))
plt.hist(mean_list, bins =  binsize, histtype = "stepfilled", alpha = 0.75, label = "Sample Means")
plt.axvline(u, linestyle = "dashed", linewidth = 2, color = "#363537", label = "True Population Mean")

plt.xlabel("Mean")
plt.ylabel("# of Sample Sets")
plt.title("Sampling and Distribution of the Mean")
plt.legend(fontsize = 8);
```

```python
# 4) mean and sd of the sampling distribution

u_dist = mean_set.mean()
sd_dist = mean_set.std(ddof = 1)

print(f"4)\nSampling Distribution Mean: {u_dist}\nSampling Distribution Standard Deviation: {sd_dist}")
```
